```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
df = pd.read_csv("insurance.csv")
df.head()
```

PRACTICAL 1

```python
#scatter plot of age against charges
x = df.iloc[:,0].values
y = df.iloc[:,6].values
y=y.reshape(-1, 1)
x=x.reshape(-1, 1)

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on the data.

#fit model
age_model = LinearRegression()
age_model.fit(x,y)

# Plot data and model with line of best fit
y_predicted = age_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_predicted,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#predict unknowns
unk_x = [[25]] 

#add unknown value to the age column
x_pred = np.append(x, unk_x).reshape(-1,1)

#predict y for the new value of x using the model
y_pred = age_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print("Insurance charges should be:", age_model.predict(unk_x))
```
